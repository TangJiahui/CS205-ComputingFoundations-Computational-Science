> **Graph (a):**
* Maximum degree of concurrency: 8

* Critical path length: 4 (any branch from grandparent to leave node)

* The minimum number of processors needed to obtain the maximum speed-up: 8

* The maximum achievable speed-up if the number of processors is limited to (i) 2 and (ii) 5:
  
  * (i) if limited to 2: It takes 8 computation unit time to finish the graph computation. 
  The maximum achievable speed-up is 
  $$ 15/8 = 1.875$$
  * (ii) if limited to 5: As critical path is 4, it is the lower bound.
  By traversing through, I found it takes at minimum 5 computation unit to finish the graph computation.
  The maximum achievable speed-up is 
  $$ 15/5 = 3$$
