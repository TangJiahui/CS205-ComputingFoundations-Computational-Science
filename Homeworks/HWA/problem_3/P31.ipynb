**Nines** given by provider as availability goals, it means uptime percentage of service per agree time unit (i.e. year/month/week etc);

For example, 
* Four nines means 99.99%;
* Five nines means 99.999%;
* 4.5 nines means 99.995% (also be called as 4N5, 4.3 OR 4.5)

For four nines, 99.99%, the maximum downtime is 365*0.01% * 24 * 60 = 52.56 minutes of down time per year.
